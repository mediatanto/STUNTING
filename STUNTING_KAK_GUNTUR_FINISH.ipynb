```python
# Step 1: Bersihkan format string aneh (ambil bagian pertama saja)
def clean_panjang(val):
    try:
        # Jika string memiliki lebih dari satu titik, ambil bagian pertama
        if isinstance(val, str) and val.count('.') >= 2:
            return val.split('.')[0]
        else:
            return val
    except:
        return np.nan
```

```python
def koreksi_berat_lahir(combined_data_satu, kolom='berat_lahir'):
    """
    Mengoreksi kesalahan input berat lahir:
    - Jika > 1000 → dianggap gram, dibagi 1000
    - Jika > 20 dan <= 1000 → dibagi 10
    """

    # 1. Bersihkan spasi dan pastikan numerik
    combined_data_satu[kolom] = combined_data_satu[kolom].astype(str).str.strip()  # hapus spasi
    combined_data_satu[kolom] = pd.to_numeric(combined_data_satu[kolom], errors='coerce')  # ubah ke float

    # 2. Buat masker koreksi satuan
    mask_bagi_1000 = combined_data_satu[kolom] > 1000
    mask_bagi_10 = (combined_data_satu[kolom] > 20) & (combined_data_satu[kolom] <= 1000)

    # 3. Log informasi koreksi
    print(f"→ Dibagi 1000 (gram → kg): {mask_bagi_1000.sum()} baris")
    print(f"→ Dibagi 10 (salah input desimal): {mask_bagi_10.sum()} baris")

    # 4. Terapkan koreksi
    combined_data_satu.loc[mask_bagi_1000, kolom] = combined_data_satu.loc[mask_bagi_1000, kolom] / 1000
    combined_data_satu.loc[mask_bagi_10, kolom] = combined_data_satu.loc[mask_bagi_10, kolom] / 10

    return combined_data_satu
```

```python
combined_data_satu = koreksi_berat_lahir(combined_data_satu, kolom='berat_lahir')
```

```python
def isi_berat_lahir_anomali(combined_data_satu, kolom='berat_lahir', min_wajar=1.0, max_wajar=5.5):
    """
    Mengganti nilai berat_lahir yang ekstrem (terlalu besar/kecil) dengan
    nilai random dari distribusi normal berdasarkan data yang wajar.
    """

    # Hitung nilai valid yang berada dalam rentang wajar
    mask_valid = (combined_data_satu[kolom] >= min_wajar) & (combined_data_satu[kolom] <= max_wajar)
    valid_values = combined_data_satu.loc[mask_valid, kolom]

    if valid_values.empty:
        print("❗ Tidak ada nilai valid untuk menghitung distribusi.")
        return combined_data_satu

    mean = valid_values.mean()
    std = valid_values.std()

    # Buat mask untuk nilai ekstrem (yang mau diganti)
    mask_extreme = (combined_data_satu[kolom] < min_wajar) | (combined_data_satu[kolom] > max_wajar)

    print(f"\n→ Nilai ekstrem yang akan diganti: {mask_extreme.sum()} baris")
    print(f"→ Distribusi pengganti ~ N({mean:.2f}, {std:.2f})")

    # Ganti nilai ekstrem dengan random normal dari distribusi valid
    np.random.seed(42)
    combined_data_satu.loc[mask_extreme, kolom] = np.random.normal(mean, std, size=mask_extreme.sum())

    return combined_data_satu
```

##MEMBUAT KOLOM TARGET

```python
def risiko_stunting(row):
    indikator_terpenuhi = 0

    # 1. Panjang lahir < 48 cm
    if row['panjang_lahir'] < 48:
        indikator_terpenuhi += 1

    # 2. Berat lahir: < 2.4 kg untuk laki-laki, < 2.5 kg untuk perempuan
    if (row['jenis_kelamin'] == 1 and row['berat_lahir'] < 2.4) or \
       (row['jenis_kelamin'] == 0 and row['berat_lahir'] < 2.5):
        indikator_terpenuhi += 1

    # 3. Usia ibu saat hamil < 20 atau > 35 tahun
    if row['usia_ibu_hamil'] < 20 or row['usia_ibu_hamil'] > 35:
        indikator_terpenuhi += 1

    # 4. Lingkar kepala lahir < 35 cm
    if row['lingkar_kepala_lahir'] < 35:
        indikator_terpenuhi += 1

    # Return 1 jika minimal 2 indikator terpenuhi, else 0
    return 1 if indikator_terpenuhi >= 2 else 0

# Terapkan ke DataFrame
combined_data_satu['risiko_stunting'] = combined_data_satu.apply(risiko_stunting, axis=1)

# Keterangan:
# 1 = Berisiko stunting
# 0 = Tidak berisiko
```

Tuning XGBoost dengan GridSearchCV

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from collections import Counter

# === 1. Siapkan data ===
X = combined_data_satu[['anak_ke', 'jenis_kelamin', 'imd','panjang_lahir','berat_lahir','lingkar_kepala_lahir','usia_ibu_hamil']]
y = combined_data_satu['risiko_stunting']

# Split data 70:30 dengan stratifikasi
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# === 2. Hitung scale_pos_weight dari data training ===
counter = Counter(y_train)
scale_pos_weight = counter[0] / counter[1]

# === 3. Setup parameter grid ===
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'scale_pos_weight': [scale_pos_weight]  # Penting untuk imbalance
}

xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='f1_macro',  # Bisa juga 'roc_auc' atau 'recall' tergantung tujuan
    cv=5,
    verbose=1,
    n_jobs=-1
)

# === 4. Latih model ===
grid_search.fit(X_train, y_train)

# === 5. Hasil terbaik ===
print("Best Parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_

# === 6. Evaluasi pada test set ===
y_pred = best_model.predict(X_test)
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
